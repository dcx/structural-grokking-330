```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
device = "cuda:0"
```

```python
model = AutoModelForCausalLM.from_pretrained("princeton-nlp/Sheared-LLaMA-1.3B", torch_dtype=torch.float16)
tokenizer = AutoTokenizer.from_pretrained("princeton-nlp/Sheared-LLaMA-1.3B")
model.to(device)
```
